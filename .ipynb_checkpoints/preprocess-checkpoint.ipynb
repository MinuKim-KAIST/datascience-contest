# Preprocessing and Observing Data
In this notebook, we aim to observe the data behavior, and choose features of importance carefully. We shall make use of functions and classes that we have written in `train.py`, `dataloader.py`, and `model.py`.

```python
import numpy as np
import pandas as pd
import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split

from dataloader import seq_collate, preprocess, exam_loader
```

```python
import os
import pathlib
import yaml

# Working directory setup
loader_root = "./loader.yml"
loader_config = yaml.load(open(loader_root, 'r'), Loader=yaml.SafeLoader)
save_dir = os.path.join('./saved_params', 'debug')
```

```python
# Feature setup
train_root_path = "./datasets/trainSet.csv"
norm_dict = loader_config['norm_factor_dict']
remove_feats = loader_config['remove_features']
```

```python
from sklearn.impute import SimpleImputer

# Setup dataset as pandas.dataframe
dataset = pd.read_csv(train_root_path)
dataset = dataset.drop(remove_feats, axis=1)

string_feats = dataset.select_dtypes(include='object').columns.tolist()
value_feats = dataset.select_dtypes(include=np.number).columns.tolist()
value_feats = [f for f in value_feats if ('MONTH' not in f and 'mo.' not in f)]

# Impute data for nan values
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

# maps the direction to an angle
direction = 'Bot-Hole direction (N/S)/(E/W)'
dataset[direction] = np.arctan(dataset[direction])

# imputes missing data
if value_feats:
    num_imputer = SimpleImputer(strategy='mean')
    dataset[value_feats] = num_imputer.fit_transform(dataset[value_feats])
if string_feats:
    cat_imputer = SimpleImputer(strategy='most_frequent')
    dataset[string_feats] = cat_imputer.fit_transform(dataset[string_feats])
    dataset = pd.get_dummies(dataset, columns=string_feats)

# Remove sequence data
total_features = [f for f in dataset.columns if 'MONTH' not in f]
dataset = dataset[total_features]

# Normalize data
for feats in norm_dict:
    dataset[feats] /= float(norm_dict[feats])
```

# Simple linear regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split, cross_val_score, KFold
import warnings
warnings.simplefilter(action='ignore')

def model_perf(model, X, Y, prefix=''):
    pred = model.predict(X)
    r2 = r2_score(Y, pred)
    rmse = mean_squared_error(Y, pred, squared=False)
    print (f'{prefix} R^2 score = {r2:.3f}, RMSE = {rmse:.4f}')
    
def model_fit_perf(model, X, Y, prefix=''):
    X_trn, X_val, Y_trn, Y_val = train_test_split(X, Y, test_size=0.2, random_state=0)
    model.fit(X_trn, Y_trn)
    print (prefix)
    model_perf(model, X_trn, Y_trn, ' Training ')
    model_perf(model, X_val, Y_val, ' Validation')
    # kfold = KFold(n_splits=5, shuffle=True, random_state=0)
    # s = cross_val_score(model, X, Y, cv=kfold)
    # print (f' Cross val. R^2 score mean = {s.mean():.3f}, std = {s.std():.3f}')
```

```python
target_name = 'Last 6 mo. Avg. GAS (Mcf)'
dataset = dataset.dropna(subset=[target_name]).reset_index(drop=True)

X = dataset[value_feats]
Y = dataset[target_name]

ridge = Ridge()
model_fit_perf(ridge, X, Y, "Ridge")
```

```python
value_dataset = dataset[value_feats]
value_dataset[target_name] = dataset[target_name]
high_corr = value_dataset.corr()[target_name].sort_values(ascending=False)
cols = high_corr[1:].keys()

XX = X.copy()
for i in cols:
    for j in cols:
        XX[f'{i} * {j}'] = X[i] * X[j]
        
ridge_corr = Ridge()
model_fit_perf(ridge_corr, XX, Y, "Ridge")
```

```python
from torch import nn

class Predictor(nn.Module):
    """
    LSTM based predictor for the next 6 month averaged gas production.
    Parameters
    ----------
    feature_dim: int
        number of static features used to predict the gas production.
    sequence_dim: int
        dimension of an element of a sequence
    hidden_dim: int
        number of features in the hidden state h
    n_layers: int
        number of recurrent layers
    """
    def __init__(self, feature_dim=22, sequence_dim=4, hidden_dim=64, n_layers=2):
        super().__init__()
        self.gru = nn.GRU(input_size=sequence_dim, hidden_size=hidden_dim,
                          num_layers=n_layers, batch_first=True)
        emb_size = feature_dim + hidden_dim
        self.fc = nn.Sequential(
            nn.Linear(emb_size, emb_size),
            nn.ReLU(),
            nn.Linear(emb_size, 1)
        )
        self.make_sos = nn.Sequential(
            nn.Linear(feature_dim, emb_size),
            nn.ReLU(),
            nn.Linear(emb_size, sequence_dim)
        )

    def forward(self, data):

        sequences, features = data

        sos = self.make_sos(features).unsqueeze(1)  # [b 1 f]
        _, hidden = self.gru(sos)
        _, hidden = self.gru(sequences.float(), hidden)

        output = torch.cat([hidden[-1], features], dim=1)
        output = self.fc(output.float())
        return output
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.dummy import DummyRegressor

X = dataset[value_feats]
Y = dataset[target_name]

ridge = Ridge()
parameters = {'alpha':[a*0.01 for a in range(400)]}
search = GridSearchCV(ridge, parameters)
search.fit(X, Y)

ridge = Ridge(**search.best_params_)
model_fit_perf(ridge, X, Y, "Ridge")

pred = ridge.predict(X)
np.set_printoptions(suppress=True)
print(f' Average L1 norm for ridge regression: {np.sum(np.abs(Y - pred))/len(pred)}')

# LSTM
print()
save_dir = os.path.join('./saved_params', '20211119020144')
model = Predictor()
model.load_state_dict(torch.load(pathlib.Path(os.path.join(save_dir, 'best_params.pkl'))))

train_file = pd.read_csv(train_root_path)
train_file = train_file.dropna(subset=[target_name]).reset_index(drop=True)
train_dataset, valid_dataset = preprocess(train_file, norm_dict, remove_feats, ratio=1)
valid_loader = DataLoader(valid_dataset, batch_size=16, shuffle=False, collate_fn=seq_collate)

predictions = []
with torch.set_grad_enabled(False):
    for _, samples in enumerate(valid_loader):
        sequences = samples['sequences']
        features = samples['features'].float()

        data = (sequences, features)
        production = model(data).view(-1).cpu().detach().numpy()
        predictions.extend(production * 1e5)

predictions = np.array(predictions)
dummy = DummyRegressor(strategy='constant', constant=predictions)
dummy.fit(X, Y)

r2 = r2_score(Y, predictions)
rmse = mean_squared_error(Y, predictions, squared=False)
print('LSTM')
print (f' Overall R^2 score = {r2:.3f}, RMSE = {rmse:.4f}')
print(f' Average L1 norm for LSTM: {np.sum(np.abs(Y - predictions))/len(predictions)}')

# kfold = KFold(n_splits=5, shuffle=True, random_state=0)
# s = cross_val_score(dummy, predictions, Y, cv=kfold)
# print (f' Cross val. R^2 score mean = {s.mean():.3f}, std = {s.std():.3f}')

# print()
# print(np.array(label - pred))
# print()
# print(np.array(label - predictions))
```

# More Classical Models

```python
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm.sklearn import LGBMRegressor

param_bagging = {'bootstrap': True, 'bootstrap_features': True, 'max_features': 0.95, 'max_samples': 0.7, 'n_estimators': 30}
param_random_forest = {'bootstrap': True, 'max_depth': 13, 'n_estimators': 30}
param_xgboost = {'learning_rate': 0.03, 'max_depth': 18, 'n_estimators': 70, 'subsample': 0.7}
param_lightgbm = {'learning_rate': 0.03, 'max_depth': 2, 'n_estimators': 190, 'subsample': 0.05}
                 
bagging = BaggingRegressor(**param_bagging)
random_forest = RandomForestRegressor(**param_random_forest)
xgb = XGBRegressor(**param_xgboost)
lgbm = LGBMRegressor(**param_lightgbm)
                 
model_fit_perf(bagging, X, Y, "Bagging")
model_fit_perf(random_forest, X, Y, "Random Forest")
model_fit_perf(xgb, X, Y, "XGBoost")
model_fit_perf(lgbm, X, Y, "LightGBM")
```
